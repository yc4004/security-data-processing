```python
import pandas as pd
import datetime
```

```python
# Read the input file
fields = []
with open(r"corp_pfd.dif", 'rb') as f:
    sign = False
    for line in f:
        l = line.rstrip().decode('utf-8')
        if l == 'START-OF-FIELDS':
            sign = True
            continue
        if l == 'END-OF-FIELDS':
            break 
        if l == '':
            continue
        if sign == True and l[0] != '#':
            fields.append(l)
            
data = []
with open(r"corp_pfd.dif", 'rb') as f:
    sign = False
    for line in f:
        if line == b'START-OF-DATA\r\n':
            sign = True
            continue
        if line == b'END-OF-DATA\r\n':
            break        
        if sign == True:
            data.append(line.rstrip().decode('utf-8').split('|')[:-1])
```

```python
# Convert it to DataFrame
df = pd.DataFrame(data, columns=fields)
df
```

```python
# Limit columns in df to only those found in reference_fields.csv
df_fields = pd.read_csv('reference_fileds.csv')
common_col = list(set(df_fields['field'].tolist()) & set(fields))
df = df[common_col]
df
```
